```python
class Car:
  def __init__(self, brand, horsepower, automatic, color, date):
    self.brand = brand
    self.horsepower = horsepower
    self.automatic = automatic
    self.color = color
    self.date = date

  def __str__(self):
    transmission = "automatic" if self.automatic else "manual"
    print_date = self.date.split("-")[0]
    return f"{self.color} {print_date} {self.brand}, {transmission} {self.horsepower}HP"



car_brand = "Porsche" # @param {type:"string"}
horsepower = 720 # @param {type:"slider", min:50, max:1200, step:5}
automatic = False # @param {type:"boolean"}
color = "yellow" # @param ["red", "blue", "green", "yellow", "black", "white"]
purchase_date = "2009-02-17" # @param {type:"date"}
my_car = Car(car_brand, horsepower, automatic, color, purchase_date)
print(my_car)
```
