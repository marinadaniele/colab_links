# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
names = ['vi', 'leen', 'alberto', 'camillo', 'joe', 'soraya', 'emma', 'edoardo']
height = [170, 180, 174, 185, 173, 166, 172, 168]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(names, height, color='purple')

# Add title and labels
plt.title('Sample Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

# Stem plot

```python
import matplotlib.pyplot as plt

names = ['vi', 'leen', 'alberto', 'camillo', 'joe', 'soraya', 'emma', 'edoardo']
height = [170, 180, 174, 185, 173, 166, 172, 168]

plt.figure(figsize=(10, 6))
plt.stem(names, height, basefmt=" ", use_line_collection=True)
plt.title('Stem Plot of Heights')
plt.xlabel('Names')
plt.ylabel('Height (cm)')
plt.show()
```

# Stairs plot

```python
import matplotlib.pyplot as plt
import numpy as np

names = ['vi', 'leen', 'alberto', 'camillo', 'joe', 'soraya', 'emma', 'edoardo']
height = [170, 180, 174, 185, 173, 166, 172, 168]

# Colors for each segment
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'cyan']

plt.figure(figsize=(10, 6))

# Plot each segment with different color using fill_between for a similar effect
for i in range(len(height)):
    plt.fill_between([i, i + 1], height[i], step='pre', color=colors[i], alpha=0.5)

plt.xticks(np.arange(len(names)), names)
plt.title('Stairs Plot of Heights')
plt.xlabel('Names')
plt.ylabel('Height (cm)')
plt.show()
```

# Fill Between Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
names = ['vi', 'leen', 'alberto', 'camillo', 'joe', 'soraya', 'emma', 'edoardo']
height_2023 = [170, 180, 174, 185, 173, 166, 172, 168]
height_2024 = [175, 178, 180, 182, 170, 165, 175, 170]

# Generate x values for the positions of the names
x = np.arange(len(names))

plt.figure(figsize=(10, 6))

# Plot the lines for two different years
plt.plot(x, height_2023, label='2023 Heights', color='blue')
plt.plot(x, height_2024, label='2024 Heights', color='green')

# Fill the area between the two lines
plt.fill_between(x, height_2023, height_2024, where=(np.array(height_2024) > np.array(height_2023)), facecolor='green', alpha=0.5, interpolate=True)
plt.fill_between(x, height_2023, height_2024, where=(np.array(height_2024) < np.array(height_2023)), facecolor='red', alpha=0.5, interpolate=True)

# Customizing the plot
plt.xticks(x, names)
plt.title('Fill Between Plot of Heights for 2023 and 2024')
plt.xlabel('Names')
plt.ylabel('Height (cm)')
plt.legend()

# Show the plot
plt.show()
```

# Stack plot

```python
import matplotlib.pyplot as plt
import numpy as np

names = ['vi', 'leen', 'alberto', 'camillo', 'joe', 'soraya', 'emma', 'edoardo']
height = [170, 180, 174, 185, 173, 166, 172, 168]

# Fake additional data for stack plot
height_1 = [h - 10 for h in height]
height_2 = [h - 20 for h in height]
height_3 = [h - 30 for h in height]

x = np.arange(len(names))

plt.figure(figsize=(10, 6))
plt.stackplot(x, height, height_1, height_2, height_3, labels=['Height', 'Height-10', 'Height-20', 'Height-30'], colors=['skyblue', 'lightgreen', 'lightcoral', 'lightgrey'])
plt.xticks(x, names)
plt.title('Stack Plot of Heights')
plt.xlabel('Names')
plt.ylabel('Height (cm)')
plt.legend(loc='upper left')
plt.show()
```
